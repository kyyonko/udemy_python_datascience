# 51. ボックスプロットとヴァイオリンプロット

```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

[http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg](http://en.wikipedia.org/wiki/Box_plot#mediaviewer/File:Boxplot_vs_PDF.svg)

```python
data1 = randn(100)
data2 = randn(100) + 2 # Off set the mean
```

```python
sns.distplot(data1)
sns.distplot(data2)
```

```python
# boxplotを作る
sns.boxplot(data = [data1, data2])
```

```python
# 外れ値を含むことが出来る
# 「ひげ」は英語で"whisker
sns.boxplot(data = [data1, data2], whis = np.inf)
```

```python
# 水平に並べることも出来る
# 引数orient
sns.boxplot(data = [data1, data2], whis = np.inf, orient = 'h')
```

```python
# Violin/Viola plot はbox plotのシンプルさとKDEの情報量を兼ね備える
```

```python
# 正規分布に従う乱数
data1 = stats.norm(0, 5).rvs(100)

# γ分布に従う乱数
data2 = np.concatenate([stats.gamma(5).rvs(50) - 1, -1 * stats.gamma(5).rvs(50)])
```

```python
# まずはボックスプロット
# ぱっと見では形変わらない
sns.boxplot(data = [data1, data2], whis = np.inf)
```

```python
#　上のをヴァイオリンプロットで表現
sns.violinplot(data = [data1, data2])
```

```python
# バンド幅を細かくする
sns.violinplot(data = data2, bw = 0.01)
```

```python
# rugプロットと同じような機能もある
sns.violinplot(data = data1, inner = "stick")
```
