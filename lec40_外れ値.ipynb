# 40. 外れ値

```python
import numpy as np
from pandas import DataFrame
```

```python
# 乱数列の種を与える
np.random.seed(12345)
dframe = DataFrame(np.random.randn(100, 4))
```

```python
dframe.head()
```

```python
dframe.tail()
```

```python
dframe.describe()
```

```python
# 最初の列
col = dframe[0]
```

```python
col.head()
```

```python
# 2より大きい要素を取り出す
col[np.abs(col) > 2]
```

```python
# DataFrameの前列に対して同じ演算をする
dframe[(np.abs(dframe) > 3).any(1)]
```

```python
np.sign(dframe)
```

```python
# 絶対値の上限が3になるようにする
dframe[np.abs(dframe) > 3] = np.sign(dframe)* 3
```
