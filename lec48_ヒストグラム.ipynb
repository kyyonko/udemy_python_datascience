```python
import numpy as np
from numpy.random import randn
import pandas as pd

from scipy import stats

import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
# 正規分布に従う乱数を作る
dataset1 = randn(100)
```

```python
# bins = 10がデフォルト
plt.hist(dataset1)
```

```python
# もう1つ作る
dataset2 = randn(80)
# 色を変える
plt.hist(dataset2, color='indianred')
```

```python
# normed	True に設定すると正規化 (合計値が 1 になるように変換) を実施。
# (デフォルト値: False)
plt.hist(dataset1, normed = True)
```

```python
# 同じ平面に正規化されたplotを重ねることができる

# normed = True として、異なる観測数のデータを比較できるようにする
# alpha = 0.5 として透明度を調整する

plt.hist(dataset1, normed = True, alpha = 0.5, bins = 20)
plt.hist(dataset2, normed = True, alpha = 0.5, bins = 20, color='indianred')
```

```python
data1 = randn(1000)
data2 = randn(1000)

# 同時分布（結合分布）
sns.jointplot(data1, data2)
```

```python
# もう少しわかりやすく
sns.jointplot(data1, data2, kind = 'hex')
```
